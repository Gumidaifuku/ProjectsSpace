```python
import modcell as mods
from IPython.display import display, HTML

# switch
switch = "1-1"
sams = "1"
titleTop = "JavaIntro"
frontNum = 1
backNum = 1
fileTitle = titleTop+str(frontNum)+'-'+str(backNum)
roopCountNum = 0
partRoopCountNum = 0
totalNum = 0
partNum = 0

while(mods._import(fileTitle,tag='title')):
    partNum = totalNum
    partRoopCountNum = roopCountNum
    
    while(mods._import(fileTitle,tag='title')):
        roopCountNum += 4
        
        for i in range(4,0,-1):
            mod = mods._import(fileTitle,tag='test'+str(i))
            if(i == 4):
                st = mod.className('demo').strD
            elif(i == 3):
                st = mod.className('demo').strC
            elif(i == 2):
                st = mod.className('demo').strB
            elif(i == 1):
                st = mod.className('demo').strA

            if(st == 'testAnswer'):
                totalNum += i
                break
        
        backNum += 1
        fileTitle = titleTop+str(frontNum)+'-'+str(backNum)
        
    partRoopCountNum = roopCountNum - partRoopCountNum
    partNum = (totalNum - partNum)/partRoopCountNum*100
    display(HTML("<h2>"+str(frontNum)+"章の進捗率</h2>"))
    display(HTML("<div style='padding: 10px; margin-bottom: 10px; border: 5px double #333333;'><h1>進捗率"+str(partNum)+"%</h1></div>"))
    
    frontNum += 1
    backNum = 1
    fileTitle = titleTop+str(frontNum)+'-'+str(backNum)
    

    
    
progress = (totalNum/roopCountNum)*100

#title
display(HTML("<h2>現在の全体進捗率</h2>"))

#進捗率表示
display(HTML("<div style='padding: 10px; margin-bottom: 10px; border: 5px double #333333;'><h1>進捗率"+str(progress)+"%</h1></div>"))
```
