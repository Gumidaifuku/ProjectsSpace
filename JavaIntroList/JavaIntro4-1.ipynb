# Java入門編 4-1 メソッドとは

**メソッドとは、次のような構文で記述します**
***
public static void メソッド名(){  
　　　命令文  
}
***

**カウントダウンメソッドを用いた例**

```java
//countdownというメソッドの名前をつけています
public static void countdown(){
    //↓ここからがカウントダウンを実行するための命令文です
        System.out.println("カウントダウンをします");
        for(int i=5 ; i>=0; i--){
            System.out.println(i);
        }
    //↑ここまでがカウントダウンを実行するための命令文です
}
//↑ここでメソッドの宣言が終了します
```
