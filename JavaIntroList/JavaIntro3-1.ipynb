**条件分岐とは**

　私たちは普段様々な選択に迫られます。所持金に応じた物の買う買わない、パスワードを入力して正しければ進み、間違いならば再入力に移る等。これらを条件に応じてコンピューターが自動で判別するため基準となるプログラムを **条件分岐** といいます。高校で数学を学んだことがある人なら真偽値という言葉を知っている人も多いと思います。その真偽値を用いるのがこの条件分岐です。

**if文**
***
if(age<20){  
　　　System.out.println("あなたは未成年です。");  
}
***
if()の()の中が条件式と呼ばれる、if文において条件分岐を行うために必要なものです。**この()の中が真である場合に{}の中が実行されます。**<br>
今回は age<20 のため、変数 age の中身が20未満である場合に真となります。

```java
# modcell: test2
# 合言葉を記入↓
answer = 'testAnswer'
    
class className():
    def __init__(self, strA):
        self.strB = answer
    
```
