**プログラムとは**

私たちが普段使っている携帯電話、ゲーム機や家電製品にはコンピューターが埋め込まれており、コンピューターがその機器ごとに仕事を行えるのは、そのコンピューターに適切な命令が送られているからです。
この命令を記したものが**プログラム**と呼ばれます。そして、コンピューターに仕事を行わせるためのプログラムを書くことを**プログラミング**と呼びます。
