# Java入門編 0-1 この教材の利用方法

## 教材の動かし方
    　この教材はJavaの記述された学習を行う場所である「JavaIntro△-□.ipynb」と進捗を確認できる「checker.ipynb」の2種類を連動させて利用します
    　cheker.ipynbの一番上のセルを実行するとファイルごとに進捗を確認できます。switch変数に△-□を入れることにより確認するファイルを変更することができます。
    　一つ下のセルを実行すると各章の進捗と全体の進捗の２つを確認することが可能です。
    　この教材はJavaの教材内に最上部が modcell ではじまるセルがあります。そこまで学習を終えたら answer 変数に testAnswer と入力しましょう
## この教材で利用するショートカットキーやコマンド一覧
    - Ctrl + Enter・・・選択したセルを実行する
    - ▶(Run the selected cells and advance)・・・選択したセルを実行する
    - Restart the kernel・・・選択しているファイルの実行状態を初期に戻す
    - 左側の青いバー・・・クリックすることで対象のセルを表示、折り畳みができる
